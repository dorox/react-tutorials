Numerical and approximated solutions for Langmuir–Hinshelwood and Eley–Rideal mechanisms
===============================

https://en.wikipedia.org/wiki/Reactions_on_surfaces


```python
from chemreact import models
```

### First example: simple decomposition

```python
sd = models.Chemistry()
sd.reaction('A+S<=>AS')
sd.reaction('AS=>P')
sd.initial_concentrations(A=1, S=0.1)
sd.run()
```

### Eley-Rideal mechanism
Single molecular adsorbtion

```python
er = models.Chemistry()
er.reaction('A+S<=>AS')
er.reaction('AS+B=>P')
er.initial_concentrations(A=1, B=1, S=0.1)
er.run()
```

### Langmuir-Hinshelwood mechanism
Bimolecular adsorbtion

```python
lh = models.Chemistry()
lh.reaction('A+S<=>AS')
lh.reaction('B+S<=>BS', k1=0.1, k2=0.1)
lh.reaction('AS+BS=>P', k=100)
lh.initial_concentrations(A=1, B=1, S=1)
lh.run()
```
